```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
a = ord('а')
alphabet = ''.join([chr(i) for i in range(a,a+32)])
ciphered = open('ciphered.txt', encoding='utf-8').read().replace('\n', '')
theor_i = 0.0553
print(ciphered)
```

```python
def get_nth_letters(text, n):
    s = ""
    for index, letter in enumerate(text):
        if index % n == 0:
            s += letter
    return s
```

```python
def count_letters(string):
    occurrences = {}
    for l in alphabet:
        occurrences[l] = string.count(l)
    return occurrences
```

```python
def count_I(text):
    n = len(text)
    s = 0
    occurrences = count_letters(text)
    for m in occurrences.values():
        s += m*(m-1)
    return s/(n*(n-1))
```

```python
# 1st method
for r in range(1, 31):
    texts = []
    for j in range(r):
        texts.append(get_nth_letters(ciphered[j:], r))
    arr_i = []
    for text in texts:
        arr_i.append(count_I(text))
    print(f'r = {r}')
    print(arr_i)
```

```python
# 2nd method
dfr = []
dfdr = []
for r in range(1, 31):
    dr = 0
    for i in range(len(ciphered) - r):
        if ciphered[i] == ciphered[i+r]:
            dr += 1
    print(f'r = {r} | Dr = {dr}')
    dfr.append(r)
    dfdr.append(dr)
```

```python
def get_max_key(d):
    max_k = 'а'
    max_v = 0
    for k in d:
        if d[k] > max_v:
            max_k = k
            max_v = d[k]
    return max_k
```

```python
def get_key_letter(l):
    return chr(ord('а') + (ord(l)-ord('о')) % len(alphabet))
```

```python
r = 12
texts = []
for j in range(r):
    texts.append(get_nth_letters(ciphered[j:], r))
for text in texts:
    occ = count_letters(text)
    n = len(text)
    for l in occ:
        occ[l] /= n
    print(get_key_letter(get_max_key(occ)), end='')
```

```python
def get_letter(ciphered_l, key_l):
    return chr(ord('а') + (ord(ciphered_l) - ord(key_l)) % len(alphabet))

def decrypt(text, key):
    deciphered = ''
    for i in range(len(text)):
        deciphered += get_letter(text[i], key[i % len(key)])
    return deciphered
```
